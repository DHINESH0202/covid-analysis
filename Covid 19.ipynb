# Relating the variables with scatterplots


```python
sns.relplot(y="TotalCases", x="TotalDeaths", hue="Continent", data=data)
```

```python
sns.pairplot(data)
```

```python
sns.relplot(y="TotalCases", x="TotalDeaths", hue="Continent",kind='line', data=data)
```

```python
sns.catplot(x="Continent", y="TotalCases", kind="bar", data=data)
```

```python
sns.catplot(x="Continent", y="TotalDeaths", kind="box", data=data)
```

```python
sns.catplot(x="Continent", y="ActiveCases", kind="strip", data=data, jitter=True)
```

```python
sns.catplot(x="Continent", y="TotalRecovered", hue="Continent", kind="violin", data=data)
```
